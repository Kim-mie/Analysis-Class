#### **Null values:**

- Missing values in a dataset
Remedy: - if the missing values are more than 80% - drop the column (delete or exclude the column from your analysis) - if the missing values are less than 20% - we drop the rows containing the missing values - if the missing values are above 40% or 50% - we conduct what we call imputation (we fill in the missing values using eg an MCT)
